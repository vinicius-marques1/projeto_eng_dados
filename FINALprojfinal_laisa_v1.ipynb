```
###########SOMA##########################
import plotly.express as px
def vacina_sum():
    dff= df_all[df_all['continent']!='NaN']
    dff = (df_all[(df_all['date'] > '2021-05-21') & (df_all['date'] < '2021-05-23')])
    temp = dff.groupby('continent')['people_fully_vaccinated', 'people_vaccinated'].sum().reset_index()
    temp=temp.melt(id_vars='continent',value_vars=['people_fully_vaccinated', 'people_vaccinated'],var_name='Status',value_name='Quantidade de pessoas')
    people_fully_vaccinated = '#21bf82'
    people_vaccinated='#A6636B'
    fig = px.bar(temp, x="continent",#y= 'people_fully_vaccinated',
                    y="Quantidade de pessoas", 
                    color='Status',
                    height=600,
                    title='Status da vacina√ß√£o contra a COVID-19 por continente nos primeiros 5 meses de 2021',
                    color_discrete_sequence = [people_fully_vaccinated,people_vaccinated])
    fig.update_layout(xaxis_rangeslider_visible=True)
    fig.show()
  
vacina_sum()
```
